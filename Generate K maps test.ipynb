```python
import numpy as np
import matplotlib.pyplot as plt

# Define the dimensions of the region
length = 2  # miles
width = 1   # mile

# Define the number of grid points
num_points_x = 100
num_points_y = 50

# Generate random temperature values for the grid
np.random.seed(0)
temperature = np.random.rand(num_points_y, num_points_x)

# Create a figure and axis
fig, ax = plt.subplots()

# Set the x and y axis limits
ax.set_xlim(0, length)
ax.set_ylim(0, width)

# Create a grid of x and y coordinates
x = np.linspace(0, length, num_points_x)
y = np.linspace(0, width, num_points_y)
X, Y = np.meshgrid(x, y)

# Plot the temperature values as a color map
cax = ax.pcolormesh(X, Y, temperature, cmap='jet')

# Define the midline
midline = length / 2

# Color the top half red
cax.set_array([])
cax.set_clim(temperature.min(), temperature.max())
cax.set_clip_path([[0, midline, length, midline, 0, 0]])

# Color the bottom half blue
cax2 = ax.pcolormesh(X, Y, temperature, cmap='jet')
cax2.set_array([])
cax2.set_clim(temperature.min(), temperature.max())
cax2.set_clip_path([[0, 0, length, midline, 0, midline]])

# Set the labels and title
ax.set_xlabel('X (miles)')
ax.set_ylabel('Y (miles)')
ax.set_title('Temperature Distribution')

# Add a colorbar
cbar = fig.colorbar(cax)

# Display the graph
plt.show()
```
