```python
import json
import numpy as np


path = "sequence_homosapien_genome_copy.json"

with open(path, "r") as file:
    data = json.load(file)
    
# Assuming data is a list of dictionaries with 'header' and 'sequences' keys
sequences_list = [entry['sequence'] for entry in data]

concatenated_sequences = ''.join(sequences_list)

unique_characters = list(set([*concatenated_sequences]))

unique_characters.sort()
    
```

```python
print(unique_characters)
```

```python

def encode_amino_acid_sequence(sequence):
    
    all_amino_acids = "ACDEFGHIKLMNPQRSTUVWY"
    amino_acid_to_int = {acid: i for i, acid in enumerate(all_amino_acids)}
    integer_sequence = [amino_acid_to_int[acid] for acid in sequence]

    # One-hot encode the sequence using lists instead of NumPy
    one_hot_encoded_sequence = [[1 if i == amino_acid_idx else 0 for i in range(len(all_amino_acids))] for amino_acid_idx in integer_sequence]

    return one_hot_encoded_sequence
```

```python
encoded_sequences = []
for i in sequences_list:
    e = encode_amino_acid_sequence(i)
    encoded_sequences.append(e)
```

```python
print(encoded_sequences[1:10])
```

```python
with open("encoded_sequences", "w") as file:
    json.dump(encoded_sequences, file, indent=4)
```
